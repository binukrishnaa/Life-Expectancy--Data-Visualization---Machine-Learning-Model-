Data Visualization

```python
import matplotlib.pyplot as plt
# importing the matplotlib.pyplot as or alias plt
plt.figure(figsize=(10,8))
# to set the figure size of the graph
plt.scatter(df['GDP_per_capita'],df['Life_expectancy'])
# a scatter plot is used to get the scatterig between the gdp per capita and life expectancy
plt.ylabel('Life Expectency')
# to the y axis 
plt.xlabel('GDP_per_captia')
# to label the x axis
plt.xscale('log')
# this function is used to log to set the value fo x in logoritm form 
plt.title('GDP vs life Expectancy')
# title is used to put the title of the graph
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#importing matplot library and seaborn library for plotting graph  
plt.figure(figsize=(35, 15))
plt.subplot(2, 2, 3)
# plotting the a larger size of the graph
Top_5=df.nlargest(5,'Alcohol_consumption')
# nlargest is a function that is used to find the largest  value among a column as specified within, where top 5 is being choosed above)
#The above are top 5 largest country with consumption of alcohol 
sns.barplot(data=Top_5,x='Country',y='Alcohol_consumption',color='red')
#plotting a bar graph using sns and give the top 5 data and comparing them
plt.title('Top five Countries by Alcohol consumption')
# plotting title of the graph
```

```python
plt.figure(figsize=(35, 15))
plt.subplot(2, 2, 3)

regional_means=df.groupby('Region')['Life_expectancy'].mean().sort_values(ascending=False)
#groupping the region and life expectancy and getting the mean value of the group values and sorting them in descending values.
regional_means.plot(kind='bar',color='orange')
#plotting the graph with the variable which the groupping is assigned
plt.xticks(rotation=5)
# xticks is used to set the lables of the graph into certain degree accordingly
plt.xlabel('REGION')
# x label is used to label the x axis
plt.ylabel('Life Expectancy')
# y label is used to label the y axis
plt.title('Life Expectany by Region')
# title is used to print the title
```

```python
top_10=df.nlargest(10,'Adult_mortality')
# using the nlargest functiom to assign the top 10 adult mortality rate to the variabel
plt.figure(figsize=(35, 15))
plt.subplot(2, 2, 3)
sns.barplot(data=top_10,x='Country',y='Adult_mortality',color='Green')
# a barplot is being used to print the barplot and using the mortality variable as data and comparing it with Country to get thr graph with color set to green
plt.xticks(rotation=10,ha='right')
# xticks is used to set the rotation of the label of the barplot accordingly
plt.title('Top ten Countries by Adult Mortality Rate')
# to print the title of the graph
plt.xlabel('Country')
# to print the label of x axis
plt.ylabel('Adult Mortality Rate')
# to print the y label of x axis
plt.tight_layout()
# its a function which is used for non overlapping of the labels among each other
```
